##### 12. Write a function using only list comprehensions, no loops, to compute standard deviation. Using this function, calculate the standard deviation of each numeric column in both data sets. Add these to a list called `sd`.

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

```python
import numpy as np
both.drop(["State"], axis = 1)
def stdv(both): #Returns the standard deviation of a numeric list
    mn = np.mean(both)
    variance = sum([(i-mn)**2 for i in both]) / len(both) 
    return variance ** .5
#the equation for standard deviation
#drinks.drop(['mL', 'servings'], axis=1, inplace=True)
```

```python
stdv(both["SAT Math"])
#testing out the standard deviation equation on a column to verify it works
```

```python
stdv(both["SAT Participation Percent"])
#testing out the standard deviation equation on another column
```

```python
sd = []

for col in both.columns:
    try:
        sd.append(stdv(both[col]))
    except:
        print('wrong data type')
sd    
#getting rid of the state so the standard deviation column can be applied to the dataset and not error out.    
```
